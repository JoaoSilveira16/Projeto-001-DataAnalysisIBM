# 4. Questões de Negócio

### Qual é a taxa geral de rotatividade de funcionários?

```python
rec_humanos['Attrition'].value_counts(normalize=True)*100
```

### O salário influencia a rotatividade de funcionários?

```python
sns.boxplot(x='Attrition', y='MonthlyIncome', data=rec_humanos)
```

### A satisfação no trabalho está relacionada à rotatividade de funcionários?

```python
sns.countplot(x='JobSatisfaction', hue='Attrition', data=rec_humanos)
```

### O equilíbrio entre vida profissional e pessoal afeta a rotatividade de funcionários?

```python
sns.countplot(x='WorkLifeBalance', hue='Attrition', data=rec_humanos)
```

# Key Insights gerados:

### 1 - A empresa apresenta uma taxa de rotatividade de aproximadamente 16%.

Isso indica que:

- a maioria dos funcionários permanece na empresa

- porém existe um nível moderado de turnover

Em empresas reais, taxas acima de 10–15% já costumam chamar atenção do RH, pois geram custos com:

- recrutamento

- treinamento

- perda de conhecimento organizacional

### 2 - Sobre o Gráfico: boxplot MonthlyIncome vs Attrition

O que aparece claramente:

- funcionários que saíram (Yes) têm salários mais baixos

- funcionários que ficaram (No) têm salários maiores em média

Portanto...

Funcionários que deixaram a empresa apresentam salários médios significativamente menores.

Isso sugere que:

- salários mais baixos podem estar associados a maior rotatividade

### 3 - Sobre o Gráfico: JobSatisfaction vs Attrition

O que observamos:

- quando JobSatisfaction = 1 ou 2, há proporcionalmente mais saídas

- quando JobSatisfaction = 3 ou 4, a maioria permanece

Ou seja:

- satisfação maior → retenção maior

Portanto...

Funcionários com baixa satisfação no trabalho apresentam maior probabilidade de deixar a empresa.

Isso indica que fatores como:

- ambiente de trabalho

- reconhecimento

- gestão

podem influenciar diretamente a retenção.

### 4 - Sobre o Gráfico: WorkLifeBalance vs Attrition

O que aparece:

- níveis 1 e 2 (pior equilíbrio) mostram mais saídas

- níveis 3 e 4 (melhor equilíbrio) mostram mais permanência

Portanto...

Funcionários com piores níveis de equilíbrio entre vida pessoal e trabalho apresentam maior tendência a sair da empresa.

Isso sugere que políticas como:

- horários flexíveis

- home office

- carga de trabalho equilibrada

podem ajudar a reduzir o turnover.

# Recomendação estratégica:
### Para reduzir a rotatividade de funcionários, a empresa deveria considerar melhorar suas estratégias de remuneração, aumentar a satisfação no trabalho e implementar políticas que promovam um melhor equilíbrio entre vida profissional e pessoal.
